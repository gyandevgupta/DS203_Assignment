```python
X = df3.drop(columns=["Rented Bike Count"])
Y = df3[["Rented Bike Count"]]
X = np.array(X)
Y = np.array(Y)
X_train, X_rem, y_train, y_rem = sklearn.model_selection.train_test_split(X, Y, test_size=0.7)
X_validate, X_test, y_validate, y_test = sklearn.model_selection.train_test_split(X_rem, y_rem, test_size=0.5)
```

##Support Vector Machine

```python
param_grid_SVM = {
    'C': [0.001,0.01,0.1, 1, 10, 100],
    'epsilon': [2,1, 0.1, 0.01, 0.001],
    'kernel': ['rbf']
}

gscv_svm = GridSearchCV(estimator = SVR(),param_grid = param_grid_SVM, cv=5)
gscv_svm_result = gscv_svm.fit(X_train,y_train.ravel())
# print(gscv_svm_result.best_estimator_)
# print(gscv_svm_result.best_score_)
```

```python
best_param_svm = gscv_svm_result.best_params_
score = cross_validate(SVR(kernel=best_param_svm['kernel'],epsilon = best_param_svm['epsilon'], C = best_param_svm['C']),X_test,y_test.ravel(),scoring = 'neg_mean_squared_error')
# print(score)
print(" RMSE :", math.sqrt(abs(score['test_score'].mean())))
```

## Random Forest

```python
param_grid_RF = {
  'n_estimators': [10, 50, 100, 200, 400, 700, 1000],
  'max_depth': np.linspace(60,130,8),
  'min_samples_leaf': [1, 2, 4]
}

gscv_rf = GridSearchCV(estimator = RandomForestRegressor(),param_grid = param_grid_RF, cv=5)
gscv_rf_result = gscv_rf.fit(X_train,y_train.ravel())
```

```python
best_param_rf = gscv_rf_result.best_params_
score = cross_validate(RandomForestRegressor(n_estimators=best_param_rf['n_estimators'],max_depth = best_param_rf['max_depth'], min_samples_leaf = best_param_rf['min_samples_leaf']),X_test,y_test.ravel(),scoring = 'neg_mean_squared_error')
# print(score)
print(" RMSE :", math.sqrt(abs(score['test_score'].mean())))
```

On comparing we get the RMSE to be lower for Random Forest so that would be a good one to apply for this dataset
